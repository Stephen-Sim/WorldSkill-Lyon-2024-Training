```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/shopping_trends_updated.csv')

colors = ['#A4C2D6', '#C75D4D', '#7A8C69']
```

```python
df_gender = df.groupby('Gender').size().sort_values(ascending=False)

df_gender.plot(kind='bar', color=colors, rot=0)

plt.xlabel("Employment Type")
plt.ylabel("Number of Occurrences")

plt.show()
```

```python
df_gender.plot(kind='pie', autopct = '%1.1f%%', colors=colors, startangle=90)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df['Age'], edgecolor='black', color="lightblue", density=True)
df['Age'].plot(kind='kde')

plt.show()
```

```python
df_cate = df.groupby('Category').size().sort_values(ascending=False)

df_cate.plot(kind='bar', color=colors, rot=0)

plt.show()
```

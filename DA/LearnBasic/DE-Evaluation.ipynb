```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

titleFontFam = font_manager.FontProperties(fname='Fonts/Montserrat-Bold.ttf')
printFontFam = font_manager.FontProperties(fname='Fonts/OpenSans-Regular.ttf')

plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = printFontFam.get_family()
colors = ['#A4C2D6', '#C75D4D', '#7A8C69']

df = pd.read_csv('Data/retail-dataset-analysis.csv', encoding='unicode_escape')

df
```

```python
# column 
# 
df.dropna(inplace=True)

df
```

```python
products = df.groupby('PRODUCTCODE').size().reset_index(name="Total Count").sort_values('Total Count', ascending=False)

best_selling_products = products[products['Total Count'] == products['Total Count'].max()]
best_selling_products.reset_index(inplace=True)

best_selling_products_code = best_selling_products['PRODUCTCODE'].tolist()

print("Best selling products: ", best_selling_products_code)
```

```python
peak_peroid = df.groupby("MONTH_ID").size().reset_index(name="Total Count").sort_values('Total Count', ascending=False).reset_index()

print("Peak peroid month: ", peak_peroid.loc[0]['MONTH_ID'])
```

```python
total_sales = df['SALES'].sum()
total_transactions = len(df)
avg_of_order_value = df['SALES'].mean()

print('total sales: ', total_sales.round(2))
print('total transactions: ', total_transactions)
print('avg of order value: ', avg_of_order_value.round(2))
```

```python
total_sales = df.groupby('MONTH_ID').agg(
    total_sales = ('SALES', 'sum'),
    total_transactions = ('ORDERNUMBER', 'count'),
    avg_sales = ('SALES', 'mean'),
).reset_index()

fig, ax = plt.subplots(3, 1, sharex= False)

ax[0].plot(total_sales['MONTH_ID'], total_sales['total_sales'], label='Total Sales', color=colors[0])
ax[1].plot(total_sales['MONTH_ID'], total_sales['total_transactions'], label='Total Transactions', color=colors[0])
ax[2].plot(total_sales['MONTH_ID'], total_sales['avg_sales'], label='Average Sales', color=colors[0])

ax[0].legend()
ax[1].legend()
ax[2].legend()

plt.savefig('Plots/Sales_vs_Transactions.png')
plt.show()
```

```python
df['Net Profit'] = df['SALES'] - (df['COSTPRICE'] * df['QUANTITYORDERED'])

product_category = df.groupby('PRODUCTCODE').agg(
    SumSales = ('SALES', 'sum'),
    SumSaleProfit = ('Net Profit', 'sum'),
).reset_index()

product_category['Profit margin'] = (product_category['SumSales'] - product_category['SumSaleProfit']) / product_category['SumSales']

product_category[['SumSales', 'Profit margin']].to_csv('Output/Profit_margin.csv', index=False)
```

```python
product_category.sort_values('SumSales', ascending=False).reset_index(inplace=True)

print('Highest profit product:', product_category.loc[0]['PRODUCTCODE'])
```

```python
product_category_sales = df.groupby('PRODUCTLINE')['SALES'].sum()

product_category_sales.plot(kind='barh', stacked=True, color=colors[0])

plt.show()
```

```python
customer_spending = df.groupby('CUSTOMERNAME')['SALES'].mean().reset_index(name="Average Spending")

customer_spending
```

```python
customer_country = df.groupby('COUNTRY').size().reset_index(name="Total Customer")

plt.barh(customer_country['COUNTRY'], customer_country['Total Customer'], color=colors[0])
plt.show()
```

```python
customer_country = df.groupby('COUNTRY')['SALES'].mean().reset_index(name="Average Spending")

fig, ax = plt.subplots(figsize=(8, 6))

plt.scatter(customer_country['COUNTRY'], customer_country['Average Spending'], color=colors[0])

plt.xticks(rotation=90)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
customer_country[['avg_t']] = scaler.fit_transform(customer_country[['Average Spending']])

def find_cluster_nums(df, max_k):
    means = []
    inertias = []
    
    for k in range(1, max_k+1):
        Kmeans = KMeans(n_clusters=k)
        Kmeans.fit(df)
        means.append(k)
        inertias.append(Kmeans.inertia_)
    
    plt.plot(means, inertias, marker='o')
    plt.show()
```

```python
find_cluster_nums(customer_country[['avg_t']], 10)
```

```python
Kmeans = KMeans(n_clusters=5)
Kmeans.fit(customer_country[['Average Spending']])

customer_country['cluster'] = Kmeans.labels_

customer_country
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for i in range(0, 5):
    data = customer_country[customer_country['cluster'] == i]
    ax.scatter(data['COUNTRY'], data['Average Spending'], label=f'cluster {i}')

plt.xticks(rotation=90)

plt.legend()
plt.show()

plt.show()
```

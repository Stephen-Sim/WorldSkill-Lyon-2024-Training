```python
import pandas as pd
import matplotlib.pyplot as plt

import ast

df = pd.read_csv('data/data_jobs.csv')

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_plot = df.groupby('job_title_short').size().sort_values(ascending=True)

df_plot.plot.barh(color="#D4A017")

plt.title("Number of Jobs per Job Titles")
plt.xlabel("Number of Jobs")

plt.show()
```

```python
df_plot = df['job_country'].value_counts().head(20).sort_values()

df_plot.plot.barh(color="#D4A017")

plt.title("Number of Jobs per Country")
plt.xlabel("Number of Jobs")
plt.ylabel("")

plt.show()
```

```python
df_plot = df['company_name'].value_counts().head(20).sort_values()

df_plot.plot.barh(color="#D4A017")

plt.title("Number of Jobs per Country")
plt.xlabel("Number of Jobs")
plt.ylabel("")

plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(10,5))

data_plot = df.groupby('job_work_from_home').size()
data_plot.plot.pie(labels=data_plot.index, autopct='%1.1f%%', startangle=90, ax=ax[0])

data_plot = df.groupby('job_no_degree_mention').size()
data_plot.plot.pie(labels=data_plot.index, autopct='%1.1f%%', startangle=90, ax=ax[1])

data_plot = df.groupby('job_health_insurance').size()
data_plot.plot.pie(labels=data_plot.index, autopct='%1.1f%%', startangle=90, ax=ax[2])

ax[0].set_title('Work From Home Offered')
ax[1].set_title('Degree Requirement')
ax[2].set_title('Health Insurance Offered')

plt.show()
```

```python
df_us = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]

df_plot = (df_us.groupby('job_location')
           .size()
           .sort_values(ascending=False)
           .head(10)
           .sort_values(ascending=True))


df_plot.plot(kind="barh", color="#D4A017")

plt.title("Counts of Job Locations for Data Analyst in the US")

plt.show()
```
